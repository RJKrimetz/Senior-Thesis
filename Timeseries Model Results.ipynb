```python
import pandas as pd
import dataframe_image as dfi
```

```python
def add_sectors(dataframe, HV = False):
    sectors = pd.read_csv('sectors.csv')
    final = dataframe.merge(sectors, on='Ticker',how='inner')
    if HV == False:
        final = final[['Ticker', 'GARCH_CI', 'GARCH_Direction', 'GARCH_RMSE','FIGARCH_CI', 'FIGARCH_Direction', 'FIGARCH_RMSE', 'APARCH_CI','APARCH_Direction', 'APARCH_RMSE', 'HARCH_CI', 'HARCH_Direction','HARCH_RMSE', 'EGARCH_CI', 'EGARCH_Direction', 'EGARCH_RMSE','Sector']]
    else:
        final = final[['Ticker', 'GARCH_CI', 'GARCH_Direction', 'GARCH_RMSE','FIGARCH_CI', 'FIGARCH_Direction', 'FIGARCH_RMSE', 'APARCH_CI','APARCH_Direction', 'APARCH_RMSE', 'HARCH_CI', 'HARCH_Direction','HARCH_RMSE','Sector']]

    return(final)

def split_sectors(data,sectors):
    num_splits = len(sectors)
    dataframes = []
    
    for i in range(0,num_splits):
        dataframes.append(data[data['Sector'] == sectors[i]])
        
    final = pd.concat(dataframes,axis = 0)
    return(final)


High_vol_sectors = ['Information Technology','Consumer Discretionary','Health Care','Energy','Communication Services']
Medium_vol_sectors = ['Materials','Industrials','Financials',]
Low_vol_sectors = ['Utilities','Real Estate','Consumer Staples']

    

#data = add_sectors(iteration_one)
#split_sectors(data,['Communication Services', 'Consumer Discretionary', 'Consumer Staples'])
```

#Two types of modes here, one is trained on Daily returns, other is trained on HV, both are looking to predict tomorrow HV

```python
daily_returns_result = pd.read_csv("Daily_Returns_Results_HV.csv")
historical_vol_result = pd.read_csv("Historical_RV_Results_HV.csv")
daily_returns_result = add_sectors(daily_returns_result)
historical_vol_result = add_sectors(historical_vol_result, HV = True) 
daily_returns_result.head()
```

```python
def summarize_results(data):
    
    list_of_results = []
    
    all_columns = data.describe()
    confidence_intervals = all_columns.iloc[:,::3]
    direction = all_columns.iloc[:,1::3]
    rmse = all_columns.iloc[:,2::3]
    
    list_of_results = [confidence_intervals,direction,rmse]
    
    
    return(list_of_results)
```

##  Daily Returns, Volatility parameter, All Stocks

```python
a = summarize_results(daily_returns_result)
#a[2].head(10)

#styled = a[2].iloc[1:,[0,1,3]]
styled = a[2].iloc[0:,[0,1,3]]
styled = styled.style.background_gradient(axis = 1)
styled.set_table_attributes("style='display:inline'").set_caption('All Tickers')
styled


#dfi.export(styled,"mytable.png")
#styled.export_png('styled.png')
#dfi.export(styled,'testing.png')
#a[1].head(10)
#a[2].head(10)
```
